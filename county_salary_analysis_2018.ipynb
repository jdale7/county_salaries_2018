```python
countysal = pd.read_csv("county_salaries_2018.csv")
```

```python
countysal.rename(columns={'Name':'name'}, inplace=True)
```

```python
countysal.rename(columns={'Employment Status':'status'}, inplace=True)
countysal.rename(columns={'Ethnic Code':'race'}, inplace=True)
countysal.rename(columns={'Gender':'sex'}, inplace=True)
countysal.rename(columns={'Department':'dept'}, inplace=True)
countysal.rename(columns={'Job Title':'job_title'}, inplace=True)
countysal.rename(columns={'Date Started':'date_started'}, inplace=True)
countysal.rename(columns={'Cur Pos Date':'current_position_start'}, inplace=True)
countysal.rename(columns={'Date Term.':'date_terminated'}, inplace=True)
countysal.rename(columns={'Pay Status':'pay_status'}, inplace=True)
countysal.rename(columns={'Annual Salary':'salary'}, inplace=True)
countysal.rename(columns={'Overtime $':'overtime'}, inplace=True)
countysal.rename(columns={'Overtime Hrs.':'overtime_hours'}, inplace=True)
countysal.rename(columns={'Overtime Rate':'overtime_rate'}, inplace=True)
countysal.rename(columns={'Bonus $':'bonus'}, inplace=True)
countysal.rename(columns={'Wellness Day':'wellness_day'}, inplace=True)
countysal.rename(columns={'Cert Bonus':'cert_bonus'}, inplace=True)
countysal.rename(columns={'Contract Bonus':'contract_bonus'}, inplace=True)
countysal.rename(columns={'Hlth Waiver':'health_waiver'}, inplace=True)
countysal.rename(columns={'Longevity':'longevity'}, inplace=True)
countysal.rename(columns={'Wellnes Bonus':'wellness_bonus'}, inplace=True)
countysal.rename(columns={'Weekend Bonus':'weekend_bonus'}, inplace=True)
countysal.rename(columns={'Gross Pay':'gross'}, inplace=True)
countysal.rename(columns={'TOTAL':'total'}, inplace=True)
```

```python
countysal.info()
```

```python
countysal['salary'] = countysal['salary'].str.replace('\$', '')
```

```python
countysal['overtime'] = countysal['overtime'].str.replace('\$', '')
countysal['overtime_rate'] = countysal['overtime_rate'].str.replace('\$', '')
countysal['bonus'] = countysal['bonus'].str.replace('\$', '')
countysal['wellness_day'] = countysal['wellness_day'].str.replace('\$', '')
countysal['cert_bonus'] = countysal['cert_bonus'].str.replace('\$', '')
countysal['contract_bonus'] = countysal['contract_bonus'].str.replace('\$', '')
countysal['health_waiver'] = countysal['health_waiver'].str.replace('\$', '')
countysal['longevity'] = countysal['longevity'].str.replace('\$', '')
countysal['wellness_bonus'] = countysal['wellness_bonus'].str.replace('\$', '')
countysal['weekend_bonus'] = countysal['weekend_bonus'].str.replace('\$', '')
countysal['gross'] = countysal['gross'].str.replace('\$', '')
countysal['total'] = countysal['total'].str.replace('\$', '')
```

```python
countysal.head()
```

```python
countysal['salary'] = countysal['salary'].str.replace('\,', '').astype(float)
countysal['overtime'] = countysal['overtime'].str.replace('\,', '').astype(float)
countysal['overtime_rate'] = countysal['overtime_rate'].str.replace('\,', '').astype(float)
countysal['bonus'] = countysal['bonus'].str.replace('\,', '').astype(float)
countysal['wellness_day'] = countysal['wellness_day'].str.replace('\,', '').astype(float)
countysal['cert_bonus'] = countysal['cert_bonus'].str.replace('\,', '').astype(float)
countysal['contract_bonus'] = countysal['contract_bonus'].str.replace('\,', '').astype(float)
countysal['health_waiver'] = countysal['health_waiver'].str.replace('\,', '').astype(float)
countysal['longevity'] = countysal['longevity'].str.replace('\,', '').astype(float)
countysal['wellness_bonus'] = countysal['wellness_bonus'].str.replace('\,', '').astype(float)
countysal['weekend_bonus'] = countysal['weekend_bonus'].str.replace('\,', '').astype(float)
countysal['gross'] = countysal['gross'].str.replace('\,', '').astype(float)
countysal['total'] = countysal['total'].str.replace('\,', '').astype(float)
```

```python
countysal['overtime_hours'] = countysal['overtime_hours'].str.replace('\,', '').astype(float)
```

```python
overtime_by_department = countysal.groupby("dept").overtime.sum().reset_index().sort_values("overtime", ascending=False)
```

```python
overtime_by_department
```

```python
ot_hours_by_dept = countysal.groupby("dept").overtime_hours.sum().reset_index().sort_values("overtime_hours", ascending=False).astype(str)
```

```python
ot_hours_by_dept
```

```python
avg_ot_rate_by_dept = countysal.groupby("dept").overtime_rate.mean().reset_index().sort_values("overtime_rate", ascending=False).astype(str)
```

```python
avg_ot_rate_by_dept
```

```python
ot_by_job = countysal.groupby("job_title").overtime.sum().reset_index().sort_values("overtime", ascending=False).astype(str)
```
