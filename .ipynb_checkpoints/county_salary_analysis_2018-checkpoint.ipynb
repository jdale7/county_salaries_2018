Setting a variable for our DataFrame:

```python
countysal = pd.read_csv("county_salaries_2018.csv")
```

```python
countysal.info()
```

Changing the column names so they are easier to work with:

```python
countysal.rename(columns={'Name':'name'}, inplace=True)
```

```python
countysal.rename(columns={'Employment Status':'status'}, inplace=True)
countysal.rename(columns={'Ethnic Code':'race'}, inplace=True)
countysal.rename(columns={'Gender':'sex'}, inplace=True)
countysal.rename(columns={'Department':'dept'}, inplace=True)
countysal.rename(columns={'Job Title':'job_title'}, inplace=True)
countysal.rename(columns={'Date Started':'date_started'}, inplace=True)
countysal.rename(columns={'Cur Pos Date':'current_position_start'}, inplace=True)
countysal.rename(columns={'Date Term.':'date_terminated'}, inplace=True)
countysal.rename(columns={'Pay Status':'pay_status'}, inplace=True)
countysal.rename(columns={'Annual Salary':'salary'}, inplace=True)
countysal.rename(columns={'Overtime $':'overtime'}, inplace=True)
countysal.rename(columns={'Overtime Hrs.':'overtime_hours'}, inplace=True)
countysal.rename(columns={'Overtime Rate':'overtime_rate'}, inplace=True)
countysal.rename(columns={'Bonus $':'bonus'}, inplace=True)
countysal.rename(columns={'Wellness Day':'wellness_day'}, inplace=True)
countysal.rename(columns={'Cert Bonus':'cert_bonus'}, inplace=True)
countysal.rename(columns={'Contract Bonus':'contract_bonus'}, inplace=True)
countysal.rename(columns={'Hlth Waiver':'health_waiver'}, inplace=True)
countysal.rename(columns={'Longevity':'longevity'}, inplace=True)
countysal.rename(columns={'Wellnes Bonus':'wellness_bonus'}, inplace=True)
countysal.rename(columns={'Weekend Bonus':'weekend_bonus'}, inplace=True)
countysal.rename(columns={'Gross Pay':'gross'}, inplace=True)
countysal.rename(columns={'TOTAL':'total'}, inplace=True)
```

What the DataFrame looks like now:

```python
countysal.head()
```

```python
countysal.info()
```

In these next few steps, we strip the '$' figures and commas from the numeric values and convert them to numbers so that we can do math on them later:

```python
countysal['salary'] = countysal['salary'].str.replace('\$', '')
```

```python
countysal['overtime'] = countysal['overtime'].str.replace('\$', '')
countysal['overtime_rate'] = countysal['overtime_rate'].str.replace('\$', '')
countysal['bonus'] = countysal['bonus'].str.replace('\$', '')
countysal['wellness_day'] = countysal['wellness_day'].str.replace('\$', '')
countysal['cert_bonus'] = countysal['cert_bonus'].str.replace('\$', '')
countysal['contract_bonus'] = countysal['contract_bonus'].str.replace('\$', '')
countysal['health_waiver'] = countysal['health_waiver'].str.replace('\$', '')
countysal['longevity'] = countysal['longevity'].str.replace('\$', '')
countysal['wellness_bonus'] = countysal['wellness_bonus'].str.replace('\$', '')
countysal['weekend_bonus'] = countysal['weekend_bonus'].str.replace('\$', '')
countysal['gross'] = countysal['gross'].str.replace('\$', '')
countysal['total'] = countysal['total'].str.replace('\$', '')
```

```python
countysal.head()
```

```python
countysal['salary'] = countysal['salary'].str.replace('\,', '').astype(float)
countysal['overtime'] = countysal['overtime'].str.replace('\,', '').astype(float)
countysal['overtime_rate'] = countysal['overtime_rate'].str.replace('\,', '').astype(float)
countysal['bonus'] = countysal['bonus'].str.replace('\,', '').astype(float)
countysal['wellness_day'] = countysal['wellness_day'].str.replace('\,', '').astype(float)
countysal['cert_bonus'] = countysal['cert_bonus'].str.replace('\,', '').astype(float)
countysal['contract_bonus'] = countysal['contract_bonus'].str.replace('\,', '').astype(float)
countysal['health_waiver'] = countysal['health_waiver'].str.replace('\,', '').astype(float)
countysal['longevity'] = countysal['longevity'].str.replace('\,', '').astype(float)
countysal['wellness_bonus'] = countysal['wellness_bonus'].str.replace('\,', '').astype(float)
countysal['weekend_bonus'] = countysal['weekend_bonus'].str.replace('\,', '').astype(float)
countysal['gross'] = countysal['gross'].str.replace('\,', '').astype(float)
countysal['total'] = countysal['total'].str.replace('\,', '').astype(float)
```

```python
countysal['overtime_hours'] = countysal['overtime_hours'].str.replace('\,', '').astype(float)
```

Now we can begin running some calculations. First, let's figure out how many people worked for the county in 2018.

```python
countysal.name.count()
```

How many of those employees worked full-time?

```python
countysal.name[countysal.status == "Full-time Regular"].count()
```

Now, we want to know how much the county paid all of its employees last year.

```python
countysal.gross.sum().round(2)
```

And how much of that total was earned by full-time workers?

```python
countysal.gross[countysal.status == "Full-time Regular"].sum().round(2)
```

Looking at the county's 2018 budget, we know it had budgeted $324,128,913 for employee salaries and wages (not including benefits like healthcare).

```python
county_salary_budget = 324128913
```

So, was the county over or under budget in terms of employee pay?

```python
(county_salary_budget - (countysal.gross.sum())).round(2)
```

It looks like the county was under budget by about $2.4 million in terms of employee pay.

Now let's breakdown the salaries, overtime, bonuses and gross pay earned in each county department. 

```python
countysal.groupby(["dept"]).salary.sum().reset_index().sort_values("salary", ascending=False).astype(str)
```

```python
countysal.groupby(["dept"]).overtime.sum().reset_index().sort_values("overtime", ascending=False).astype(str)
```

```python
countysal.groupby(["dept"]).bonus.sum().reset_index().sort_values("bonus", ascending=False).astype(str)
```

```python
countysal.groupby(["dept"]).gross.sum().reset_index().sort_values("gross", ascending=False).astype(str)
```

Below, we create a variable to view the total overtime earned by employees in each county department:

```python
overtime_by_department = countysal.groupby("dept").overtime.sum().reset_index().sort_values("overtime", ascending=False)
```

```python
overtime_by_department
```

Now let's examine the gender and race breakdown of the county workforce. 

```python
countysal.groupby("race").name.count().reset_index().sort_values("name", ascending=False)
```

```python
countysal.groupby("sex").name.count().reset_index().sort_values("name", ascending=False)
```

From these tables we can see that the county workforce is pretty evenly split between men and women, though white people make up a significant majority of the employees. Let's now break this data down further by men and women separately.

```python
countysal.groupby(countysal.race[countysal.sex == "Male"]).sex.count().reset_index().sort_values("sex", ascending=False)
```

Now let's look at the median salaries for both men and women, broken down by race. The table below shows the median salaires of men of all races. 

```python
countysal.groupby(countysal.race[countysal.sex == "Male"]).salary.median().reset_index().sort_values("salary", ascending=False)
```

The table below shows the the median salaires of women of all races. 

```python
countysal.groupby(countysal.race[countysal.sex == "Female"]).salary.median().reset_index().sort_values("salary", ascending=False)
```

Now let's make the same tables, but looking at gross pay rather than regular salaries.

```python
countysal.groupby(countysal.race[countysal.sex == "Male"]).gross.median().reset_index().sort_values("gross", ascending=False)
```

```python
countysal.groupby(countysal.race[countysal.sex == "Female"]).gross.median().reset_index().sort_values("gross", ascending=False)
```

Now let's take a closer look at overtime earned by county workers. First we'll look at the total overtime hours worked by employees in each county department.

```python
ot_hours_by_dept = countysal.groupby("dept").overtime_hours.sum().reset_index().sort_values("overtime_hours", ascending=False).astype(str)
```

```python
ot_hours_by_dept
```

In the table below, we take a look at the average overtime rate earned by workers in each department.

```python
avg_ot_rate_by_dept = countysal.groupby("dept").overtime_rate.mean().reset_index().sort_values("overtime_rate", ascending=False).astype(str)
```

```python
avg_ot_rate_by_dept
```

In the table below we can see the total overtime worked by people in each position, listed from greatest to least.

```python
ot_by_job = countysal.groupby("job_title").overtime.sum().reset_index().sort_values("overtime", ascending=False).astype(str)
```

```python
ot_by_job
```

In the table below, we can see the average overtime hours worked by people in each position, listed from greatest to least.

```python
countysal.groupby("job_title").overtime_hours.mean().reset_index().sort_values("overtime_hours", ascending=False).astype(str)
```

The table below shows all county departments and how much overtime was worked collectively in each.

```python
countysal.groupby(["dept"]).overtime_hours.sum()
```

This next table shows which departments paid their workers the most overtime and also how many overtime hours those employees worked.

```python
overtime_table = pd.pivot_table(countysal,index=["dept"],values=["overtime", "overtime_hours"],aggfunc=np.sum).sort_values("overtime", ascending=False).astype(str).round(2)
```

```python
overtime_table
```
